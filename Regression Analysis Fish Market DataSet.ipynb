```python
Weight_Height=pd.DataFrame(fish_data,columns=['Height','Weight'])
x=Weight_Height.iloc[:,:-1].values
y=Weight_Height['Weight'].tolist()
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=1/2,random_state=0)
```

```python
from sklearn.linear_model import LinearRegression
lin=LinearRegression()
```

```python
lin.fit(x_train,y_train)
```

```python
print('Intercept: \n',lin.intercept_)
print('Coefficients: \n',lin.coef_)
```

```python
y_pred=lin.predict(x_test)
```

```python
from sklearn import metrics
print('mean squared error: %.2f'%np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
```

```python
%matplotlib inline

# Enlarge the plot size
plt.figure(figsize=(20,10))


plt.plot(x_test,y_test,'o',color='red',markersize = 14,label = "Weight vs Height training set")
plt.plot(x_test,y_pred, '-',color='black',label = 'Simple Linear Regression Line', linewidth = 4)

# y coordinate for centroid
y_mean = y_pred.mean()    
# x coordinate for centroid
x_mean = x_test.mean()
print ('The centroid for this dataset is x = {:.2f} and y = {:.2f}.'.format(x_mean, y_mean))
# Plot the centroid point
plt.plot(x_mean,y_mean, '*', markersize = 50, color = "b") 

# Create legend
plt.legend(loc = 'upper right', fontsize = 20)

# Enlarge x and y tick marks
plt.xticks(fontsize = 20)
plt.yticks(fontsize = 20)
# Create the x and y axis labels
plt.xlabel('Height', fontsize = 30)
plt.ylabel('Weight', fontsize = 30)
plt.title('Weight=f(Height)',fontsize = 30)
plt.grid(True)
plt.show()
```
