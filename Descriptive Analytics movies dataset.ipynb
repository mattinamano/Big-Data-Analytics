```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("companies_sorted.csv") 
```

```python
data
```

```python
company=pd.DataFrame(data)
```

```python
company=company[pd.notnull(company['country'])]
```

```python
company
```

```python
company.dropna()
```

```python
country = company['country'].value_counts(dropna=True, sort=False).nlargest(10)
country_freq = pd.DataFrame(country)
country_freq = country_freq.reset_index()
country_freq.columns = ['country', 'freq']
country_freq.sort_values(by=['freq'], inplace=True)
plt.xticks(rotation=90)
plt.yticks(np.arange(0, 10000000, 250000))
plt.title('Frequency of companies in countries') 
plt.xlabel('country') 
plt.ylabel('Frequency') 
plt.plot(country_freq['country'],country_freq['freq'],'r',linewidth=0.8)
```

```python
country = company['country'].value_counts(dropna=True, sort=False).nlargest(10)
country_freq = pd.DataFrame(country)
country_freq = country_freq.reset_index()
country_freq.columns = ['country', 'freq']
country_freq.sort_values(by=['freq'], inplace=True)
plt.xticks(rotation=90)
plt.yticks(np.arange(0, 10000000, 150000))
plt.title('Frequency of companies in countries')
plt.xlabel('country') 
plt.ylabel('Frequency') 
plt.bar(country_freq['country'],country_freq['freq'],color = '#00BFFF')
```

```python
country = company['country'].value_counts(dropna=True, sort=False).nlargest(10)
country_freq = pd.DataFrame(country)
country_freq = country_freq.reset_index()
country_freq.columns = ['country', 'freq']
country_freq.sort_values(by=['freq'], inplace=True)
plt.rcParams['font.size'] = 10.0
plt.axis('equal')
plt.pie(country_freq['freq'], labels = country_freq['country'],autopct='%1.2f%%',wedgeprops={"edgecolor":"black"})
plt.legend(loc='upper center', bbox_to_anchor=(1.45, 0.8), shadow=True, ncol=1)
plt.show()
```

```python
employee_cnt = company['current employee estimate'].value_counts(dropna=True, sort=False)
employee_cnt
```

```python
unwanted_zeroes_employee_cnt=company[company['current employee estimate'].astype(int)==0].index
unwanted_zeroes_employee_cnt
```

```python
company.drop(unwanted_zeroes_employee_cnt, inplace=True)
```

```python
# company=company[company['current employee estimate']>2]
# company
emp_company_est=company[company['current employee estimate']>2]
emp_company_est
```

```python
employee_cnt = emp_company_est['current employee estimate'].value_counts(dropna=True, sort=False)
emp_cnt = pd.DataFrame(employee_cnt)
emp_cnt = emp_cnt.reset_index()
emp_cnt.columns = ['current employee estimate', 'freq']
emp_cnt.sort_values(by=['freq'], inplace=True)
emp_cnt=emp_cnt[emp_cnt['freq']>10000]
plt.xticks(rotation=90)
plt.title('Frequency of employee estimate across countries') 
plt.xlabel('current employee estimate') 
plt.ylabel('Frequency') 
plt.plot(emp_cnt['current employee estimate'],emp_cnt['freq'],'r',linewidth=0.8)
```
