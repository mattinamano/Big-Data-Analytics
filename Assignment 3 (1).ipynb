```python
data1 = pd.read_csv("ratings.csv") 
ratings=pd.DataFrame(data1)
```

```python
ratings
```

```python
unwanted_zeroes_rating=ratings[ratings['rating'].astype(int)==0].index
unwanted_zeroes_rating
```

```python
ratings.drop(unwanted_zeroes_rating, inplace=True)
```

```python
ratings
```

```python
rating_dist = ratings['rating'].value_counts(dropna=True, sort=False)
rating = pd.DataFrame(rating_dist)
rating = rating.reset_index()
rating.columns = ['rating', 'freq']
rating.sort_values(by=['rating'], inplace=True)
plt.title('Frequency of rating') 
plt.xlabel('Rating') 
plt.ylabel('Frequency') 
plt.plot(rating['rating'],rating['freq'],'r',linewidth=0.8)
```

```python
rating_dist = ratings['rating'].value_counts(dropna=True, sort=False)
rating = pd.DataFrame(rating_dist)
rating = rating.reset_index()
rating.columns = ['rating', 'freq']
#rating.sort_values(by=['rating'], inplace=True)
plt.title('Frequency of rating') 
plt.xlabel('Rating') 
plt.ylabel('Frequency') 
plt.xticks(np.arange(1, 5.5, 0.5))
plt.bar(rating['rating'],rating['freq'],edgecolor="black",color = '#800080')
```

```python
rating_dist = ratings['rating'].value_counts(dropna=True, sort=False)
rating = pd.DataFrame(rating_dist)
rating = rating.reset_index()
rating.columns = ['rating', 'freq']
rating.sort_values(by=['rating'], inplace=True)
plt.rcParams['font.size'] = 10.0
plt.axis('equal')
plt.pie(rating['freq'], labels = rating['rating'],autopct='%1.2f%%',wedgeprops={"edgecolor":"black"})
plt.legend(labels = rating['rating'],loc=5)
plt.show()
```
